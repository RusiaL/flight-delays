```python
# Missing values
missing_counts = df.isnull().sum()
missing_percent = (missing_counts / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Count': missing_counts,
    'Missing Percent': missing_percent
}).sort_values(by='Missing Percent', ascending=False)

print("üîç Columns with missing values:")
display(missing_df[missing_df['Missing Count'] > 0])
```

```python
essential_cols = ['ARRIVAL_DELAY', 'DEPARTURE_DELAY', 'SCHEDULED_DEPARTURE']
df_cleaned = df.dropna(subset=essential_cols)

num_cols = df_cleaned.select_dtypes(include=['float64', 'int64']).columns
df_cleaned[num_cols] = df_cleaned[num_cols].fillna(df_cleaned[num_cols].median())

# cat_cols = df_cleaned.select_dtypes(include=['object']).columns
# for col in cat_cols:
#     df_cleaned[col] = df_cleaned[col].fillna(df_cleaned[col].mode()[0])

print("‚úÖ Remaining missing values after cleaning:")
print(df_cleaned.isnull().sum())
```
